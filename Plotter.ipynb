# Combining the fits using derived analytical formula

```python
covMat0=np.cov(amp1_delay,amp2_delay)
covMat=np.linalg.inv(covMat0)
```

```python
print(covMat0)
print(covMat)
```

```python
A0=[]
for i in range(len(amp1_delay)):
    num1=(covMat[0][0]+covMat[0][1])*amp1_delay[i]
    num2=(covMat[1][0]+covMat[1][1])*amp2_delay[i]
    num= num1+num2
    den=covMat[0][0]+(2*covMat[0][1])+covMat[1][1]
    A0.append(num/den)
```

```python
#plt.hist(A0,bins=100, range=(0.0028,0.0032), alpha=0.4, label= "Best fit")
plt.hist(amp32768_delay,bins=100, range=(0.0028,0.0032), alpha=0.4, label= "amp32768")
plt.hist(amp1_delay,bins=100, range=(0.0028,0.0032), alpha=0.4, label= "amp1")
plt.hist(amp2_delay,bins=100, range=(0.0028,0.0032), alpha=0.4, label= "amp2")
plt.legend()
plt.show()
```
